```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
df = pd.read_csv('../data/cleaned_albania_youth_data.csv')
```

```python
# NEET trend
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='year', y='neet_rate', marker='o')
plt.title("NEET Rate in Albania (2016â€“2023)")
plt.savefig('../data/images/neet_rate_trend.png')
plt.show()
```

```python
# Youth emigration bar plot
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='year', y='youth_emigration', color="skyblue")
plt.title("Youth Emigration per Year")
plt.savefig('../data/images/emigration_per_year.png')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.savefig('../data/images/correlation_heatmap.png')
plt.show()
```

```python
# Dual axis: Wage vs Emigration
fig, ax1 = plt.subplots(figsize=(8,5))
ax1.set_xlabel("Year")
ax1.set_ylabel("Youth Emigration", color="red")
ax1.plot(df['year'], df['youth_emigration'], color="red", marker='o')

ax2 = ax1.twinx()
ax2.set_ylabel("Average Wage", color="blue")
ax2.plot(df['year'], df['avg_gross_wage'], color="blue", marker='o')

plt.title("Wages vs Youth Emigration Over Time")
plt.savefig('../data/images/emigration_vs_wage.png')
plt.show()
```
